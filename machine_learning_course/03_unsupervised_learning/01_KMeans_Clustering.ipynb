## 📚 THEORY

**Unsupervised Learning** = No labels. Find hidden patterns in data.

### K-Means Algorithm
```
1. Pick K random points as cluster centers (centroids)
2. Assign each point to the nearest centroid
3. Move centroids to the mean of their assigned points
4. Repeat steps 2-3 until centroids stop moving
```

### How to choose K? — The Elbow Method
- Plot **inertia** (sum of squared distances to centroids) vs K
- Pick the K where the curve bends (the 'elbow')

### Key Terms
| Term | Meaning |
|------|---------|
| Inertia | Total distance of points from their centroid (lower = better) |
| Silhouette Score | How similar a point is to its cluster vs others (higher = better, max 1) |

## 🔬 PRACTICAL

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs, load_iris
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
print('Libraries loaded ✅')
````

### Practical 1: K-Means Step by Step

````python
# Generate toy data with 3 natural clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.8, random_state=42)

# Train K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# Plot: Before vs After clustering
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(X[:, 0], X[:, 1], color='gray', alpha=0.5, s=20)
axes[0].set_title('Before Clustering (no labels)')

axes[1].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=20, alpha=0.7)
axes[1].scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')
axes[1].set_title('After K-Means (K=3)')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f'Inertia: {kmeans.inertia_:.2f}')
print(f'Silhouette Score: {silhouette_score(X, labels):.3f}')
````

### Practical 2: Elbow Method — Choosing the Right K

````python
inertias = []
silhouettes = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X, km.labels_))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(K_range, inertias, 'bo-')
axes[0].axvline(3, color='red', linestyle='--', label='Elbow at K=3')
axes[0].set_title('Elbow Method')
axes[0].set_xlabel('K')
axes[0].set_ylabel('Inertia')
axes[0].legend()

axes[1].plot(K_range, silhouettes, 'go-')
axes[1].axvline(K_range[np.argmax(silhouettes)], color='red', linestyle='--', label='Best K')
axes[1].set_title('Silhouette Score')
axes[1].set_xlabel('K')
axes[1].set_ylabel('Silhouette Score')
axes[1].legend()

plt.tight_layout()
plt.show()
````

### Practical 3: K-Means on Real Data (Iris)

````python
iris = load_iris()
X_iris = StandardScaler().fit_transform(iris.data)

km_iris = KMeans(n_clusters=3, random_state=42, n_init=10)
km_iris.fit(X_iris)
predicted_clusters = km_iris.labels_

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# True labels
axes[0].scatter(X_iris[:, 0], X_iris[:, 1], c=iris.target, cmap='Set1', s=30, alpha=0.8)
axes[0].set_title('True Species Labels')
axes[0].set_xlabel('Sepal Length (scaled)')

# Predicted clusters
axes[1].scatter(X_iris[:, 0], X_iris[:, 1], c=predicted_clusters, cmap='Set1', s=30, alpha=0.8)
axes[1].set_title('K-Means Predicted Clusters')
axes[1].set_xlabel('Sepal Length (scaled)')

plt.tight_layout()
plt.show()

print(f'Silhouette Score: {silhouette_score(X_iris, predicted_clusters):.3f}')
print('\n💡 K-Means recovered species groupings without seeing any labels!')
````

## 🎓 Summary

| Concept | Takeaway |
|---------|----------|
| K-Means | Assign points to nearest centroid, move centroids, repeat |
| Inertia | Lower is better — sum of distances to centroids |
| Elbow Method | Find K where inertia stops dropping sharply |
| Silhouette Score | Closer to 1 = better separation |
| Always scale! | K-Means uses distances — feature scale matters |
